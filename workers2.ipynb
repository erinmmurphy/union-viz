```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
ws_um = pd.read_csv('/Users/erinmurphy/Documents/GitHub/union-viz/data/um_ws_2008-22.csv')
ws_um.head()
```

```python
ws_um.info()
```

```python
# seaborn combined line plot of u_mem_pct and ws_pct over time
sns.lineplot(x="year", y="ws_pct", data=ws_um)
sns.lineplot(x="year", y="u_mem_pct", data=ws_um)

# set label
plt.ylabel("numbers")

plt.show()

```

```python
# seaborn combined bar plot of u_mem_pct and ws_pct over time
sns.barplot(x="year", y="ws_pct", data=ws_um)
sns.barplot(x="year", y="u_mem_pct", data=ws_um)

# set label
plt.ylabel("numbers")

plt.show()
```
